Configure the dataset for performance

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

Standardizing and Normalizing the data

```python
resize_and_rescale = tf.keras.Sequential([
  layers.experimental.preprocessing.Resizing(256, 256),
  layers.experimental.preprocessing.Rescaling(1./255),
])
```

The Sequential model consists of three convolution blocks (tf.keras.layers.Conv2D) with a max pooling layer (tf.keras.layers.MaxPooling2D) in each of them. There's a fully-connected layer (tf.keras.layers.Dense) with 128 units on top of it that is activated by a ReLU activation function ('relu'). 

```python
input_shape = (32, 256, 256, 3)
n_classes = 2
model = tf.keras.Sequential([
    resize_and_rescale,
    
    layers.BatchNormalization(),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),

    Conv2D(filters=32, kernel_size=(2,2), activation='relu', input_shape=(256, 256, 3)),
    MaxPooling2D(),
    
    Conv2D(filters=64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(),
    Dropout(0.3), 
    
    Conv2D(filters=64, kernel_size=(4,4), activation='relu'),
    Conv2D(filters=128, kernel_size=(5,5), activation='relu'),
    MaxPooling2D(),
    Dropout(0.4),
    
    Conv2D(filters=128, kernel_size=(5,5), activation='relu'),
    MaxPooling2D(),
    Dropout(0.5),
    
    Flatten(), 
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dense(3, activation='softmax')
])
model.build(input_shape=input_shape)
```

```python
model.summary()
```

 To view training and validation accuracy for each training epoch

Calculating if theres any loss

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_ds,
    batch_size=32,
    validation_data=val_ds,
    verbose=1,
    epochs=40,
)
```

```python
scores = model.evaluate(test_ds)
```
